## Section 2: Preprocessing Data

### Theory

Raw features often have very different scales ŌĆö tumour radius might range 6ŌĆō28 mm while smoothness ranges 0.05ŌĆō0.16. Many algorithms (especially SVM, logistic regression, and PCA) are sensitive to scale. **Standardization** fixes this by transforming each feature to zero mean and unit variance:

```
z = (x - mean) / std
```

In sklearn:
```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)   # fit on train, transform train
X_test_scaled = scaler.transform(X_test)   # ONLY transform test ŌĆö never fit on test!
```

**PCA (Principal Component Analysis)** reduces high-dimensional data to fewer dimensions while preserving the most variance. Useful for visualization (2D plot) and removing noise.

```python
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_scaled)  # always apply PCA on scaled data
print(pca.explained_variance_ratio_)  # how much variance each component explains
```

## Section 3: Creating and Evaluating Models

### Theory

Every sklearn estimator follows the same three-method API ŌĆö learn it once, use it everywhere:

```python
model = LogisticRegression()   # 1. Instantiate with hyperparameters
model.fit(X_train, y_train)    # 2. Train on training data
y_pred = model.predict(X_test) # 3. Predict on new data
```

**Key evaluation metrics**:

| Metric | When to use |
|--------|-------------|
| `accuracy_score` | Balanced classes |
| `f1_score` | Imbalanced classes ŌĆö harmonic mean of precision & recall |
| `classification_report` | Full breakdown per class |
| `confusion_matrix` | See exactly where the model goes wrong |

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

acc = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred, target_names=["neg", "pos"]))
cm  = confusion_matrix(y_test, y_pred)
```

## Section 4: Train-Test Split

### Theory

The golden rule of ML evaluation: **never evaluate on data you trained on**. A train-test split reserves a held-out portion of data the model never sees during training.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% for testing, 80% for training
    random_state=42,      # seed for reproducibility
    stratify=y            # keep class proportions the same in both splits
)
```

**Why `stratify=y`?** If your dataset has 90% class A and 10% class B, a random split might put all class B samples in train and none in test. `stratify=y` ensures both splits mirror the original class ratios.

```python
# Check that class ratios are preserved
import pandas as pd
print(pd.Series(y_train).value_counts(normalize=True))
print(pd.Series(y_test).value_counts(normalize=True))
# Both should show the same percentages
```

**Common split ratios**:
- 80/20 ŌĆö standard, plenty of training data
- 70/30 ŌĆö when you have limited data
- 60/20/20 ŌĆö train/validation/test for hyperparameter tuning

### Demo: Splitting and verifying stratification

````python
# ----- DEMO: Stratified split and verification -----

# Use the make_classification synthetic dataset ŌĆö fully controllable
X_syn, y_syn = make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=5,
    n_classes=3,           # three classes, deliberately imbalanced
    weights=[0.6, 0.3, 0.1],  # 60% class 0, 30% class 1, 10% class 2
    random_state=SEED
)

print(f"Full dataset: {X_syn.shape}")
print(f"Class distribution in full set:")
for cls, cnt in zip(*np.unique(y_syn, return_counts=True)):
    print(f"  class {cls}: {cnt} ({cnt/len(y_syn)*100:.1f}%)")

# Stratified 70/30 split
X_tr_syn, X_te_syn, y_tr_syn, y_te_syn = train_test_split(
    X_syn, y_syn,
    test_size=0.30,
    random_state=SEED,
    stratify=y_syn   # <-- this is the key
)

print(f"\nTrain: {len(X_tr_syn)} samples  |  Test: {len(X_te_syn)} samples")
print("\nClass distribution TRAIN:")
for cls, cnt in zip(*np.unique(y_tr_syn, return_counts=True)):
    print(f"  class {cls}: {cnt} ({cnt/len(y_tr_syn)*100:.1f}%)")
print("\nClass distribution TEST:")
for cls, cnt in zip(*np.unique(y_te_syn, return_counts=True)):
    print(f"  class {cls}: {cnt} ({cnt/len(y_te_syn)*100:.1f}%)")
# Both should mirror ~60/30/10 ŌĆö that's stratification working
````

## Section 5: Cross-Validation

### Theory

A single train-test split gives you **one** accuracy number. Did you get lucky with a particularly easy test set? Cross-validation solves this by rotating the held-out fold across the whole dataset.

**k-fold cross-validation** (k=5 is the default):

```
Fold 1: [TEST] [train] [train] [train] [train]
Fold 2: [train] [TEST] [train] [train] [train]
Fold 3: [train] [train] [TEST] [train] [train]
...
```

You get 5 accuracy scores. The mean and std together are a much more reliable estimate of true performance.

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

# Simple: 5-fold CV returning accuracy scores
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
print(f"Mean: {scores.mean():.4f}  Std: {scores.std():.4f}")

# Better: use StratifiedKFold to maintain class proportions in each fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y, cv=skf, scoring='f1_macro')
```

**Common `scoring` values**: `'accuracy'`, `'f1_macro'`, `'f1_weighted'`, `'roc_auc'`

### Demo: 5-fold stratified CV on iris

````python
# ----- DEMO: 5-fold stratified CV on iris -----

# Model to evaluate
lr_cv_demo = LogisticRegression(max_iter=200, random_state=SEED)

# StratifiedKFold maintains class proportions across folds
skf_demo = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

# cross_val_score trains and evaluates the model on each fold
cv_scores = cross_val_score(
    lr_cv_demo,         # estimator
    X_iris_scaled,      # full feature matrix (CV handles the splitting)
    y_iris,             # full labels
    cv=skf_demo,        # cross-validation strategy
    scoring="accuracy"  # metric
)

print("Scores per fold:", np.round(cv_scores, 4))
print(f"\nMean accuracy : {cv_scores.mean():.4f}")
print(f"Std  accuracy : {cv_scores.std():.4f}")
print(f"95% CI        : [{cv_scores.mean()-2*cv_scores.std():.4f}, "
      f"{cv_scores.mean()+2*cv_scores.std():.4f}]")
# The std tells you how stable the model is across different data subsets
````

## Section 6: Hyperparameter Tuning

### Theory

Most ML models have **hyperparameters** ŌĆö settings you choose before training (not learned from data). Examples:

| Model | Key hyperparameters |
|-------|---------------------|
| `LogisticRegression` | `C` (regularisation strength), `solver` |
| `RandomForest` | `n_estimators`, `max_depth`, `min_samples_split` |
| `SVC` | `C`, `kernel`, `gamma` |

**GridSearchCV** exhaustively tries every combination of hyperparameters using cross-validation:

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear']
}

gs = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid,
    cv=5,                    # 5-fold CV for each combination
    scoring='f1_macro',      # metric to optimise
    n_jobs=-1                # use all CPU cores
)
gs.fit(X_train, y_train)

print(gs.best_params_)   # best hyperparameter dict
print(gs.best_score_)    # best CV score
best_model = gs.best_estimator_  # already fitted with best params
```

**Warning**: GridSearchCV can be slow. If you have 4 values ├Ś 2 values ├Ś 5 folds = 40 model fits. With large datasets and many combinations, consider `RandomizedSearchCV` instead.

### Demo: GridSearchCV on the breast cancer dataset

````python
# ----- DEMO: GridSearchCV for LogisticRegression on breast cancer -----

# Parameter grid ŌĆö 4 x 2 = 8 combinations, each tried with 5-fold CV = 40 fits
param_grid_demo = {
    'C': [0.01, 0.1, 1.0, 10.0],    # regularisation: smaller C = more regularisation
    'solver': ['lbfgs', 'liblinear'] # optimisation algorithm
}

gs_demo = GridSearchCV(
    LogisticRegression(max_iter=2000, random_state=SEED),
    param_grid_demo,
    cv=5,                 # 5-fold stratified CV (default for classifiers)
    scoring='f1_macro',   # optimise for macro F1
    n_jobs=-1,            # parallelise across all CPU cores
    return_train_score=True
)

# Fit on the full scaled dataset (CV does internal splitting)
gs_demo.fit(X_bc_scaled, y_bc)

print(f"Best parameters : {gs_demo.best_params_}")
print(f"Best CV F1      : {gs_demo.best_score_:.4f}")
print(f"\nAll results sorted by mean test score:")

import pandas as pd
results_df = pd.DataFrame(gs_demo.cv_results_)
cols = ['param_C', 'param_solver', 'mean_test_score', 'std_test_score', 'rank_test_score']
print(results_df[cols].sort_values('rank_test_score').to_string(index=False))
````
